```python
import pandas as pd
from faker import Faker
import json
import random
import psycopg2
from psycopg2.extras import register_uuid
```

```python
fake = Faker('ru_RU')
```

```python
def generate_users(user_id):
    gender = "M" if random.randint(0, 1) else "F"
    users = {
        "user_id": user_id,
        "fname": fake.first_name_male() if "M" else fake.first_name_female(),
        "sname": fake.last_name_male() if "M" else fake.last_name_female(),
        "email": fake.email(),
        "phone": fake.phone_number(),
        "address": fake.address(),
        "gender": gender,
        "age": random.randint(17, 45),
        "is_suspicious": 0 == random.randint(0, 1),
    }
    return users

def generate_couriers(courier_id):
    сouriers = {
        "courier_id": courier_id,
        "fname": fake.first_name_male(),
        "sname": fake.last_name_male(),
        "phone": fake.phone_number(),
        "rating": round(random.uniform(2, 5), 2),
        "gender": "M",
        "age": random.randint(23, 54)
    }
    return сouriers

def generate_orders(order_id):
    orders = {
        "order_id": order_id,
        "product_ids": [random.randint(1, 100) for _ in range(random.randint(1, 5))],
        "creation_date": fake.date_between(start_date='-20y').strftime('%Y-%m-%d'),
    }
    return orders

def generate_products(product_id, product_names):
    products = {
        "product_id": product_id + 1,
        "pname": product_names[product_id],
        "price": random.randint(150, 4500),
    }
    return products

def generate_courier_vehicle():
    vehicles = []
    for courier_id in range(1, 100001):
        
        vehicle = {
            "courier_id": courier_id,
            "vehicle_id": courier_id + 10,
        }
        vehicles.append(vehicle)
        
    return vehicles
```

```python
users = [generate_users(user_id) for user_id in range(1, 500001)]

couriers = [generate_couriers(courier_id) for courier_id in range(1, 100001)]

orders = [generate_orders(order_id) for order_id in range(1, 1000001)]

products = [generate_products(product_id, product_names) for product_id in range(len(product_names))]

vehicles = generate_courier_vehicle()
```

```python
def generate_user_actions():
    user_actions = []
    
    for order_id in range(1, 1000001):
        user_id = random.randint(1, 500000)
        payment_type = "card" if random.randint(0, 1) else "cash"
        
        user_action = {
            "user_id": user_id,
            "action": "create_order",
            "order_id": order_id,
            "payment_type": payment_type,
        }

        user_actions.append(user_action)
            
        if random.randint(0, 8) >= 6:
            user_action = {
                "user_id": user_id,
                "action": "cancel_order",
                "order_id": order_id,
                "payment_type": payment_type,
            }
            
            user_actions.append(user_action)
            
    return user_actions

def generate_courier_actions():
    courier_actions = []
    
    for order_id in range(1, 1000001):
        courier_id = random.randint(1, 100000)
           
        courier_action = {
            "courier_id": courier_id,
            "action": "accept_order",
            "order_id": order_id,
        }
        
        courier_actions.append(courier_action)
        
        if random.randint(0, 8) < 6:
            courier_action = {
                "courier_id": courier_id,
                "action": "deliver_order",
                "order_id": order_id,
            }
            courier_actions.append(courier_action)
    return courier_actions
```

```python
user_actions = generate_user_actions()

courier_actions = generate_courier_actions()
```

```python
connection = psycopg2.connect(user="postgres",
                              password="admin",
                              host="localhost",
                              port="5432",
                              database="Delivery_service")

cursor = connection.cursor()
print("Информация о сервере PostgreSQL")
print(connection.get_dsn_parameters(), "\n")
cursor.execute("SELECT version();")
record = cursor.fetchone()
print("Вы подключены к - ", record, "\n")
```

```python
connection.rollback()

for user in users:
    cursor.execute('INSERT INTO public."Users" (user_id, fname, sname, gender, age, address, phone, email, is_suspicious) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)',
                (user['user_id'], user['fname'], user['sname'], user['gender'], user['age'], user['address'], user['phone'], user['email'], user['is_suspicious']))

for courier in couriers:
    cursor.execute('INSERT INTO public."Couriers" (courier_id, fname, sname, phone, rating, gender, age) VALUES (%s, %s, %s, %s, %s, %s, %s)',
                (courier['courier_id'], courier['fname'], courier['sname'], courier['phone'], courier['rating'], courier['gender'], courier['age']))

for order in orders:
    cursor.execute('INSERT INTO public."Orders" (order_id, product_ids, creation_date) VALUES (%s, %s, %s)',
                (order['order_id'], order['product_ids'], order['creation_date']))
    
for product in products:
    cursor.execute('INSERT INTO public."Products" (product_id, pname, price) VALUES (%s, %s, %s)',
                (product['product_id'], product['pname'], product['price']))

for vehicle in vehicles:
    cursor.execute('INSERT INTO public."courier_vehicle" VALUES (%s, %s)',
                (vehicle['courier_id'], vehicle['vehicle_id']))

for user_action in user_actions:
    cursor.execute('INSERT INTO public."User_actions" (user_id, action, order_id, payment_type) VALUES (%s, %s, %s, %s)',
                (user_action['user_id'], user_action['action'], user_action['order_id'], user_action['payment_type']))
    
for courier_action in courier_actions:
    cursor.execute('INSERT INTO public."Courier_actions" (courier_id, action, order_id) VALUES (%s, %s, %s)',
                (courier_action['courier_id'], courier_action['action'], courier_action['order_id']))

connection.commit()
cursor.close()
```
